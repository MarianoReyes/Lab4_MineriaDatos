# Secci√≥n 1

```python
import pandas as pd
import matplotlib.pyplot as plt

# cargar el conjunto de datos iris.csv
data = pd.read_csv('iris.csv')

# visualizar los primeros 5 registros del conjunto de datos
print(data.head())
```

### Una vez tenemos el archivo leido procedemos a crear los clusters

```python
from sklearn.cluster import KMeans

# crear un objeto KMeans con 2 clusters
kmeans = KMeans(n_clusters=2)

# entrenar el modelo de clustering
kmeans.fit(data[['sepal_length', 'sepal_width']])

# asignar una etiqueta de cluster a cada punto
labels = kmeans.predict(data[['sepal_length', 'sepal_width']])

# crear un gr√°fico de dispersi√≥n de longitud del s√©palo versus ancho del s√©palo, colorizado por etiqueta de cluster
plt.scatter(data['sepal_length'], data['sepal_width'], c=labels)
plt.xlabel('Longitud del S√©palo')
plt.ylabel('Ancho del S√©palo')
plt.show()
```

### Como siguiente paso estandarizaremos los datos y volveremos a hacer el paso 2

```python
from sklearn.preprocessing import StandardScaler

# estandarizar los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])

# crear un objeto KMeans con 2 clusters
kmeans = KMeans(n_clusters=2)

# entrenar el modelo de clustering
kmeans.fit(scaled_data)

# asignar una etiqueta de cluster a cada punto
labels = kmeans.predict(scaled_data)

# crear un gr√°fico de dispersi√≥n de longitud del s√©palo versus ancho del s√©palo, colorizado por etiqueta de cluster
plt.scatter(data['sepal_length'], data['sepal_width'], c=labels)
plt.xlabel('Longitud del S√©palo')
plt.ylabel('Ancho del S√©palo')
plt.show()
```

### Usamos el m√©todo del codo para conocer los clusters ideales

```python
# calcular la suma de las distancias al cuadrado para diferentes valores de k
sum_of_squared_distances = []
K = range(1,11)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled_data)
    sum_of_squared_distances.append(kmeans.inertia_)

# graficar la suma de las distancias al cuadrado versus k
plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Suma de las distancias al cuadrado')
plt.title('M√©todo del codo para encontrar el n√∫mero √≥ptimo de clusters')
plt.show()
```

### Creamos diferentes gr√°ficas bas√°ndonos en el n√∫mero de clusters que se ajusten mejor a los datos

```python
# crear un objeto KMeans con 2 clusters
kmeans = KMeans(n_clusters=2)

# entrenar el modelo de clustering
kmeans.fit(scaled_data)

# asignar una etiqueta de cluster a cada punto
labels = kmeans.predict(scaled_data)

# crear un gr√°fico de dispersi√≥n de longitud del s√©palo versus ancho del s√©palo, colorizado por etiqueta de cluster
plt.scatter(data['sepal_length'], data['sepal_width'], c=labels)
plt.xlabel('Longitud del S√©palo')
plt.ylabel('Ancho del S√©palo')
plt.title('2 clusters')
plt.show()

# crear un objeto KMeans con 3 clusters
kmeans = KMeans(n_clusters=3)

# entrenar el modelo de clustering
kmeans.fit(scaled_data)

# asignar una etiqueta de cluster a cada punto
labels = kmeans.predict(scaled_data)

# crear un gr√°fico de dispersi√≥n de longitud del s√©palo versus ancho del s√©palo, colorizado por etiqueta de cluster
plt.scatter(data['sepal_length'], data['sepal_width'], c=labels)
plt.xlabel('Longitud del S√©palo')
plt.ylabel('Ancho del S√©palo')
plt.title('3 clusters')
plt.show()

# crear un objeto KMeans con 4 clusters
kmeans = KMeans(n_clusters=4)

# entrenar el modelo de clustering
kmeans.fit(scaled_data)

# asignar una etiqueta de cluster a cada punto
labels = kmeans.predict(scaled_data)

# crear un gr√°fico de dispersi√≥n de longitud del s√©palo versus ancho del s√©palo, colorizado por etiqueta de cluster
plt.scatter(data['sepal_length'], data['sepal_width'], c=labels)
plt.xlabel('Longitud del S√©palo')
plt.ylabel('Ancho del S√©palo')
plt.title('4 clusters')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from kneed import KneeLocator
import matplotlib.pyplot as plt

# cargar datos
iris = pd.read_csv("iris.csv")

# estandarizar datos
iris_std = (iris - iris.mean()) / iris.std()

# calcular SSE para diferentes valores de k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(iris_std)
    sse.append(kmeans.inertia_)

# encontrar punto de codo
kl = KneeLocator(range(1, 11), sse, curve="convex", direction="decreasing")
knee = kl.elbow

# graficar la suma de la distancia al cuadrado vs k y el punto de codo
plt.plot(range(1, 11), sse, marker="o")
plt.xlabel("k")
plt.ylabel("Suma de la distancia al cuadrado")
plt.vlines(knee, plt.ylim()[0], plt.ylim()[1], linestyles="--")
plt.annotate("Punto de Codo", xy=(knee, sse[knee-1]), xytext=(4, 100), textcoords="offset points", fontsize=10, arrowprops=dict(facecolor="black", arrowstyle="->"))
plt.show()

print(f"El n√∫mero √≥ptimo de clusters encontrado con knee es: {knee}")
```

Con esto podemos concluir que el n√∫mero encontrado con el m√©todo del codo fue el acertado y el m√©todo de clustering se realiz√≥ correctamente :D !
